## Rember our wordcount code
you could import it here, but I repeated it to act as a reminder

```python
def wordcount(userwords):

    wordlist=userwords.split()

    wordcount = dict()
    for word in wordlist:
        if word in wordcount:
            wordcount[word] += 1
        else:
            wordcount[word] = 1
    return wordcount

def a_union_b(a,b):
    union = list()
    keysa=a.keys()
    keysb=b.keys()
    for word in keysa:
        if word in keysb:
#            print("found :" + word)
            union.append(word)
    return union


def in_a_not_in_b(a, b):
    a_not_b = list()
    keysa=a.keys()
    keysb=b.keys()
    for word in keysa:
        if word not in keysb:
#            print("count not find :" + word)
            a_not_b.append(word)
#        else:
#            print("found :" + word)
    return a_not_b
```
